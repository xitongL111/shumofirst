### 1. 问题的假设与符号定义

- 设企业从供应商处采购一批零配件，这批零配件的次品率为 $p$。
- 供应商声称次品率不超过标称值 $p_0 = 0.1$。
- 企业进行抽样检验，样本大小为 $n$，在抽样过程中次品的数量为 $X$。
- 次品的数量 $X$ 服从二项分布：

$$ X \sim \text{Binomial}(n, p) $$

其中，$p$ 是实际次品率，$n$ 是抽样数量。

为了简化问题，在样本数量较大时可以使用**正态分布近似**，即：

$$ X \sim N(np, np(1 - p)) $$

我们接下来建立两个情景下的数学模型。

### 2. 情景一：95% 信度下认定次品率超过标称值，拒收零配件

#### 2.1 假设检验设定

我们要检验是否在 95% 的信度下认定次品率超过标称值 $p_0 = 0.1$，即需要进行假设检验：

- **零假设**：$H_0: p = p_0 = 0.1$
- **备择假设**：$H_1: p > p_0$

这是一个**单侧检验**，即我们假设实际次品率大于标称值 $p_0$。

#### 2.2 正态近似

根据**大数定理**，当样本量 $n$ 足够大时，二项分布可以近似为正态分布：

$$ X \sim N(np_0, np_0(1 - p_0)) $$

其中，$X$ 是次品的数量，$n$ 是抽样的样本数，$p_0 = 0.1$ 是标称次品率。

为了在 95% 的信度下进行检验，查标准正态分布表可以得出显著性水平 $\alpha = 0.05$ 对应的 $Z$ 值为 $Z_{0.05} = 1.645$。于是我们可以构建标准化统计量：

$$ Z = \frac{X - np_0}{\sqrt{np_0(1 - p_0)}} $$

根据检验标准，当 $Z > 1.645$ 时，我们可以拒绝零假设 $H_0$，即认定实际次品率 $p$ 超过标称值 $p_0$。

#### 2.3 样本量 $n$ 的计算

为确保有足够的检验功效，我们需要计算出合适的样本量 $n$。根据正态分布的假设：

$$ X = np_0 + Z \cdot \sqrt{np_0(1 - p_0)} $$

将 $Z = 1.645$ 代入：

$$ X = np_0 + 1.645 \cdot \sqrt{np_0(1 - p_0)} $$

我们可以根据此公式来确定 $X$ 值（即临界次品数），并通过实际检测结果与 $X$ 比较，决定是否拒收。

### 3. 情景二：90% 信度下认定次品率不超过标称值，接收零配件

#### 3.1 假设检验设定

在这个情景中，我们要检验在 90% 的信度下认定次品率不超过标称值 $p_0 = 0.1$，即进行假设检验：

- **零假设**：$H_0: p = p_0 = 0.1$
- **备择假设**：$H_1: p \leq p_0$

这也是一个**单侧检验**，但方向相反，我们希望认定实际次品率不超过标称值。

#### 3.2 正态近似

同样使用正态分布近似：

$$ X \sim N(np_0, np_0(1 - p_0)) $$

在 90% 信度下，显著性水平 $\alpha = 0.10$，对应的 $Z$ 值为 $Z_{0.10} = 1.28$。标准化统计量为：

$$ Z = \frac{X - np_0}{\sqrt{np_0(1 - p_0)}} $$

当 $Z < 1.28$ 时，接受零假设 $H_0$，即认定次品率不超过标称值 $p_0 = 0.1$。

#### 3.3 样本量 $n$ 的计算

类似于情景一，我们可以计算临界次品数：

$$ X = np_0 + Z \cdot \sqrt{np_0(1 - p_0)} $$

将 $Z = 1.28$ 代入，可以得到 $X$ 的值。通过与实际检测结果对比，确定是否接收零配件。

### 4. 模型总结

我们设计的抽样检测方案基于**假设检验**，其中次品数量 $X$ 服从二项分布，在样本量足够大时近似为正态分布。通过不同信度下的假设检验，我们可以设定拒收和接收零配件的标准。

- **情景一（拒收）**：当次品率超过标称值，在 95% 信度下，使用 $Z = 1.645$ 作为检验标准，设计抽样方案。
- **情景二（接收）**：当次品率不超过标称值，在 90% 信度下，使用 $Z = 1.28$ 作为检验标准，设计抽样方案。

根据样本量 $n$ 和临界次品数 $X$，我们可以判断是否接收该批零配件。
